Schreiben Sie eine Methode `twinPrimes` welche zwei ganzzahlige Parameter `lo` und `hi` entgegennimmt und alle ganzen Zahlen $n$ zwischen `lo` und `hi` (Grenzen inklusive) ausgibt, für die sowohl $n$ als auch $n+2$ prim sind. (Für uns ist 1 keine Primzahl). Beim Aufruf `twinPrimes(2, 11)` müsste die Methode also die Zahlen 3, 5 und 11 ausgeben.

````java
class Numbers {

    // Ihre Lösung

    public static void main(String[] args) {
        // Aufruf für lo = 2 und hi = 11
        twinPrimes(2, 11);
    }

}
````

````java
Numbers.main(new String[0]);
````

#### Lösung

````java
class Numbers {

    public static boolean isPrime(int n) {
        if (n == 1) {return false;}
        for (int i = 2; i <= n/2; i++) {
            if (n % i == 0) {return false;}
        }
        return true;
    }

    public static void twinPrimes(int lo, int hi) {
        for (int i = lo; i <= hi; i++) {
            if (isPrime(i) && isPrime(i + 2)) {
                System.out.println(i);
            }
        }
    }

    public static void main(String[] args) {
        // Aufruf für lo = 2 und hi = 11
        twinPrimes(2, 11);
    }

}
````

Falls Sie `String.format` verwenden können Sie das Formatierungszeichen `%s` für Strings verwenden. 

````java
class Gap {
    
    // Ihre Lösung
    
    public static void main(String[] args) {
        System.out.println(makeGap("Java", "Fun", 1));
        System.out.println(makeGap("Java", "Fun", 2));
    }
}
````

#### Lösung

````java
class Gap {
    
    public static String makeGap(String a, String b, int gap) {
        return String.format(a + "%" + gap + "s" + b, "");
    }
    
    public static void main(String[] args) {
        System.out.println(makeGap("Java", "Fun", 1));
        System.out.println(makeGap("Java", "Fun", 2));
    }
}
````

## Aufgabe 5 (4 Punkte)
Schreiben Sie eine Methode `drawPyramid` welche eine positive ganze Zahl `n` entgegennimmt und eine Pyramide von Höhe `n` ausgibt. Eine Pyramide besteht hier aus Sternen `*`, die so angeordnet sind, dass in jeder höheren Schicht sowohl links als auch recht ein Stern weniger steht. Für $n=4$ sollten Sie die folgende Pyramide erhalten:

```
   *   
  ***  
 ***** 
*******
```

````java
class Pyramid {

    // Ihre Lösung

    public static void main(String[] args) {
        drawPyramid(5);
    }

}
````

````java
Pyramid.main(new String[0]);
````

#### Lösung

````java
class Pyramid {

    public static void drawPyramid(int n) {
        for (int i = 0; i < n; i++) {
            for (int j = 0; j < n - i - 1; j++) {
                System.out.print(" ");
            }
            for (int j = n - i - 1; j < n + i; j++) {
                System.out.print("*");
            }
            for (int j = n + i; j < 2 * n - 1; j++) {
                System.out.print(" ");
            }
            System.out.println("");
        }
    }

    public static void main(String[] args) {
        drawPyramid(5);
    }

}
````

````java
class Recursion {

    public static int sumOfSquares(int n) {
        if (n <= 0) {return 0;}
        else {
            return n*n + sumOfSquares(n-1);
        }
    }

    public static void main(String[] args) {
        System.out.println(sumOfSquares(5));
    }
}
````

````java
class Turns{

    // Diese Methode sollen Sie nicht verändern
    public static boolean cross(double x1, double y1, double x2, double y2, double x3, double y3, double x4, double y4) {
        if (Math.signum(((y2 - y1) * (x3 - x2)) - ((x2 - x1) * (y3 - y2))) != Math.signum(((y2 - y1) * (x4 - x2)) - ((x2 - x1) * (y4 - y2))) &&
            Math.signum(((y4 - y3) * (x2 - x4)) - ((x4 - x3) * (y2 - y4))) != Math.signum(((y4 - y3) * (x1 - x4)) - ((x4 - x3) * (y1 - y4)))) {
            return true;
        }
        return false;
    }

    public static boolean thrackle(double[][] points) {
        for (int i = 0; i < points.length - 3; i++) {
            if (!cross(points[i][0], points[i][1], points[i+1][0], points[i+1][1], points[i+2][0], points[i+2][1], points[i+3][0], points[i+3][1])) {
                return false;
            }
        }
        return true;
    }

    public static void main(String[] args) {
        double[] p1 = new double[]{-1.72, -0.18};
        double[] p2 = new double[]{1.84, 3.68};
        double[] p3 = new double[]{-2.38, 3.98};
        double[] p4 = new double[]{1.66, -0.48};
        double[] p5 = new double[]{0.08, 5.38};
        double[] p6 = new double[]{-0.32, 0.36};
        double[] p7 = new double[]{3.44, 2.74};

        double[][] points = {p1, p2, p3, p4, p5, p6, p7};
        System.out.println(thrackle(points));
    }
}
````
